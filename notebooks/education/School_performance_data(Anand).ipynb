```python
import pandas as pd

# List of file paths to your CSV files
file_paths = [
    "../raw_data/school_performance/tx_rend_municipios_2008.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2009.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2010.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2011.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2012.csv",
    "../raw_data/school_performance//tx_rend_municipios_2013.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2014.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2015.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2016.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2017.csv",
    "../raw_data/school_performance//tx_rend_municipios_2018.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2019.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2020.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2021.csv", 
    "../raw_data/school_performance//tx_rend_municipios_2022.csv",

]

# Read each CSV and store in a list of DataFrames
dfs = [pd.read_csv(file_path) for file_path in file_paths]

# Concatenate all DataFrames together
combined_df = pd.concat(dfs, ignore_index=True)

```

```python
combined_df
```

```python
columns_to_drop = ['Região', 'UF', 'Nome do Município', 'Dependência Administrativa']
combined_df = combined_df.drop(columns=columns_to_drop)
```

```python
combined_df
```

```python
combined_df = combined_df.rename(columns={
    'Código do Município': 'Código_IBGE'
})
combined_df
```

```python
# Write the combined DataFrame to a new CSV file
file_path = "../transformed_data/school_performance_per_municipality.csv"
combined_df.to_csv(file_path, index=False)
```
