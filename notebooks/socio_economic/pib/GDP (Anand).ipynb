```python
import pandas as pd
```

```python
pib_df = pd.read_csv("../raw_data/PIB/updated_pib2000-2020.csv", sep=";")
pib_df.head()
```

```python
# Generating the list of years using list comprehension
years = [str(year) for year in range(2000, 2021)]

# Melting the DataFrame
melted_df = pd.melt(pib_df, id_vars=["Estado","Codigo UF","UF", "Código_IBGE","Região"], value_vars=years, var_name="Ano", value_name='PIB')


# Sorting the DataFrame
melted_df = melted_df.sort_values(by=["Estado","Codigo UF","UF", "Código_IBGE","Região", "Ano"]).reset_index(drop=True)

melted_df
```

```python
melted_df = melted_df.drop(columns="Código_IBGE")
melted_df
```

```python
melted_df['PIB'] = melted_df['PIB'].str.replace(',','.').astype(float)
```

```python
melted_df
```

```python
file_path = "../transformed_data/gdp_per_capita_data.csv"
melted_df.to_csv(file_path, index=False)
```
