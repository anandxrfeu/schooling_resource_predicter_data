```python
# List of file paths to your CSV files
file_paths = [
    '../raw_data/Desnutrição/Desnutricao 2008.csv',
    '../raw_data/Desnutrição/Desnutricao 2009.csv',
    '../raw_data/Desnutrição/Desnutricao 2010.csv',
    '../raw_data/Desnutrição/Desnutricao 2011.csv',
    '../raw_data/Desnutrição/Desnutricao 2012.csv',
    '../raw_data/Desnutrição/Desnutricao 2013.csv',
    '../raw_data/Desnutrição/Desnutricao 2014.csv',
    '../raw_data/Desnutrição/Desnutricao 2015.csv',
    '../raw_data/Desnutrição/Desnutricao 2016.csv',
    '../raw_data/Desnutrição/Desnutricao 2017.csv',
    '../raw_data/Desnutrição/Desnutricao 2018.csv',
    '../raw_data/Desnutrição/Desnutricao 2019.csv',
    '../raw_data/Desnutrição/Desnutricao 2020.csv',
    '../raw_data/Desnutrição/Desnutricao 2021.csv',
    '../raw_data/Desnutrição/Desnutricao 2022.csv',
    '../raw_data/Desnutrição/Desnutricao 2023.csv'
]
```

```python
dfs = [pd.read_csv(file_path) for file_path in file_paths]
```

```python
# Concatenate all DataFrames together
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
# Write the combined DataFrame to a new CSV file
file_path = "../transformed_data/updated_desnutricao.csv"
combined_df.to_csv(file_path, index=False)
```
