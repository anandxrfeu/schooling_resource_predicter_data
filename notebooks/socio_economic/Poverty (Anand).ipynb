```python
import pandas as pd
```

```python
poverty_df = pd.read_csv("../transformed_data/Poverty_rate_updated- 2012-2021.csv", sep=";")
poverty_df.head()
```

```python
poverty_df = poverty_df.drop(columns=["Código_IBGE", "2021", "2020", "2019", "2018", "2017", "2016", "2015", "2014", "2013", "2012"])
poverty_df.head()
```

```python
# Raname columns
poverty_df = poverty_df.rename(columns={
    'Menos_de_US$_5,5_%_2021': '2021',
    'Menos_de_US$_5,5_%_2020': '2020',
    'Menos_de_US$_5,5_%_2019': '2019',
    'Menos_de_US$_5,5_%_2018': '2018',
    'Menos_de_US$_5,5_%_2017': '2017',
    'Menos_de_US$_5,5_%_2016': '2016',
    'Menos_de_US$_5,5_%_2015': '2015',
    'Menos_de_US$_5,5_%_2014': '2014',
    'Menos_de_US$_5,5_%_2013': '2013',
    'Menos_de_US$_5,5_%_2012': '2012',
})
poverty_df
```

```python
# Generating the list of years using list comprehension
years = [str(year) for year in range(2012, 2022)]

# Melting the DataFrame
melted_df = pd.melt(poverty_df, id_vars=["Estado", "Código_UF", "UF"], value_vars=years, var_name='Ano', value_name='Poverty_Rate')


# Sorting the DataFrame
melted_df = melted_df.sort_values(by=["Estado", "Código_UF", "UF","Ano"]).reset_index(drop=True)

melted_df
```

```python
melted_df['Poverty_Rate'] = melted_df['Poverty_Rate'].str.replace(',', '.', regex=True).astype(float)
melted_df
```

```python
file_path = "../transformed_data/poverty_data.csv"
melted_df.to_csv(file_path, index=False)
```
