```python
import pandas as pd
```

```python
malnutrition_df = pd.read_csv("../../../transformed_data/socio_economic/desnutrição/updated_desnutricao.csv", sep=";")
malnutrition_df.head()
```

```python
malnutrition_df = malnutrition_df.drop(columns=["Magreza Acentuada", "Magreza", "Eutrofia","Eutofia %","Sobrepeso","Soprepeso %","Obesidade","Obesidade %","Obesidade Grave","Obesidade Grave %","Total"])
```

```python
malnutrition_df
```

```python
malnutrition_df['Magreza Acentuada %']  = malnutrition_df['Magreza Acentuada %'] .replace('-', '0%')
malnutrition_df['Magreza %']  = malnutrition_df['Magreza %'] .replace('-', '0%')

malnutrition_df['Magreza Acentuada %'] = malnutrition_df['Magreza Acentuada %'].str.rstrip('%').astype(float)
malnutrition_df['Magreza %'] = malnutrition_df['Magreza %'].str.rstrip('%').astype(float)

malnutrition_df["Magreza_total_%"] = malnutrition_df['Magreza Acentuada %'] + malnutrition_df['Magreza %']
```

```python
malnutrition_df = malnutrition_df.drop(columns=["Magreza Acentuada %", "Magreza %", "Região", "Código UF", "UF", "Municipio"])
```

```python
malnutrition_df
```

```python
file_path = "../../../transformed_data/socio_economic/desnutrição/malnutrition_per_municipality_ML.csv"
malnutrition_df.to_csv(file_path, index=False)
```
