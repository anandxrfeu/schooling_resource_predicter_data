```python
import pandas as pd
```

```python
unemployement_df = pd.read_csv("../../../raw_data/socio_economic/unemployment/desocupacao_updated.csv", sep=";")
unemployement_df.head()
```

```python
unemployement_df = unemployement_df.drop(columns=["UF", "Código_IBGE", "Estado", "Região"])
unemployement_df
```

```python
# Generating the list of years using list comprehension
years = [str(year) for year in range(2012, 2023)]

# Melting the DataFrame
melted_df = pd.melt(unemployement_df, id_vars=["Código_UF"], value_vars=years, var_name='Ano', value_name='Unemployed_%')


# Sorting the DataFrame
melted_df = melted_df.sort_values(by=["Código_UF"]).reset_index(drop=True)

melted_df
```

```python
melted_df['Unemployed_%'] = melted_df['Unemployed_%'].astype(str).str.replace(',', '.').astype(float)
```

```python
file_path = "../../../transformed_data/socio_economic/unemployment/unemployement_per_UF_ML.csv"
melted_df.to_csv(file_path, index=False)
```
