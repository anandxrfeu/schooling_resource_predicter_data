```python
unemployement_df = pd.read_csv("../transformed_data/desocupacao_updated.csv", sep=";")
```

```python
unemployement_df.head()
```

```python
unemployement_df = unemployement_df.drop(columns="Código_IBGE")
```

```python
unemployement_df
```

```python
# Generating the list of years using list comprehension
years = [str(year) for year in range(2012, 2023)]

# Melting the DataFrame
melted_df = pd.melt(unemployement_df, id_vars=["UF", "Código_UF","Estado","Região"], value_vars=years, var_name='Ano', value_name='Unemployed')


# Sorting the DataFrame
melted_df = melted_df.sort_values(by=["UF", "Código_UF","Estado","Região","Ano"]).reset_index(drop=True)

melted_df
```

```python
file_path = "../transformed_data/unemployement_data.csv"
melted_df.to_csv(file_path, index=False)
```
