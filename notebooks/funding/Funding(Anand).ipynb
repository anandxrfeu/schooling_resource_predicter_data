```python
import pandas as pd
```

```python
df = pd.read_csv("../raw_data/funding/transferências_para_municípios.csv", sep=";",  encoding="ISO-8859-1")
df
```

```python
#Remove 'R$' and convert 'Valor Consolidado' column to numeric
df['Valor Consolidado'] = df['Valor Consolidado'].str.replace('R$', '', regex=False)
df['Valor Consolidado'] = df['Valor Consolidado'].str.replace('.', '', regex=False)
df['Valor Consolidado'] = df['Valor Consolidado'].str.replace(',', '.', regex=True).astype(float)
```

```python
df
```

```python
# Group by 'Município', 'Ano', 'UF' and 'Código IBGE' and sum 'Valor Consolidado'
df_grouped = df.groupby(['Município', 'Ano', 'UF', 'Código IBGE'])['Valor Consolidado'].sum().reset_index()
df_grouped
```

```python
columns_to_drop = ['Município', 'UF']
df = df_grouped.drop(columns=columns_to_drop)
df
```

```python
# Export the cleaned and pivoted dataframe to CSV
file_path = "../transformed_data/funding_per_municipality.csv"
df.to_csv(file_path, index=False)
```
